```python
# Read in tab seperated tsv file
cities = pd.read_csv("../tsv/cities_canada-usa.tsv", sep='\t')
```

```python
# Draw a sample to get a visual on data
cities.head()
```

```python
# Get some information about the table
cities.info()
```

```python
# 1. For the suggestions feature, we only need a subset of the loaded dataframe
cities_subset = cities[['name', 
             'ascii',
             'lat',
             'long',
             'country' ]]
```

```python
# Draw a sample to get a visual on data
cities_subset.head()
```

```python
# Get some information about the table
cities_subset.info()
```

```python
# We are to drop any rows
cities_subset.dropna(thresh=1)
```

```python
# Convert all the values to strings to simplyfy processing
cities_subset = cities_subset.applymap(str)
```

```python
# Convert to json files in persistance
cities_subset.to_json('../json/processed-cities.json', 
                               orient = "records",
                               date_format = "epoch",
                               double_precision = 10,
                               force_ascii = True,
                               date_unit = "ms",
                               default_handler = None)
```
